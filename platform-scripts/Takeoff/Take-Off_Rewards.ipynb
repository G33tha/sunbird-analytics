```python
import json
data = []
with open('DeltaDump2015All.txt') as f:
    for line in f:
        data.append(json.loads(line))
d=dict()
for item in data:
    if(item['uid']!="none" and item['uid'] not in d.keys()):
        ud=item['uid']
        c="No"
        for time in data:
            if(time['uid']==ud and'id' in time['edata']['eks'].keys()):
                s=time['edata']['eks']['id']
                if("suitcase" in s):
                    c="Yes"
                    break
        d[ud]=c
print "{:<38}{:<15}".format('UID', "Access rewards Page")
for k, v in d.iteritems():
    print "{:<38}{:<15}".format(k, v)
```
