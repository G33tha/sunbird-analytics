```python
%matplotlib inline
import json
import matplotlib.pyplot as plt
count=0
c=0
td=dict()
data = []
with open('DeltaDump2015All.txt') as f:
    for line in f:
        data.append(json.loads(line))
d=dict()
for item in data:
    if(item['uid']!="none" and item['uid'] not in d.keys()):
        ud=item['uid']
        t=list()
        for time in data:
            if(time['uid']==ud):
                t.append(time['ts'])
        sorted(t)
        s=int((t[0])[11:13])
        td[s]=td.get(s,0)+1
        d[ud]=t[0]
plt.title("Time of Day Users are on Take Off")
plt.xlabel("Time of day")
plt.ylabel("No. of Users")
plt.axis([0, 24, 0, 20])
plt.plot(td.keys(), td.values())
plt.show()
```
